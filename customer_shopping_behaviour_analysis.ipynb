```python
# replacing null values
df.groupby('Category')['Review Rating'].mean()
```

```python
df.groupby('Category')['Review Rating'].median()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# changing the column names in snake casing for consistancy
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns= {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group

labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(30)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# creating a column purchase_frequency_days

frequency_mapping = {
    'Daily': 1,
    'Weekly': 7,
    'Fortnightly':14,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).fillna(0)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(20)
```
